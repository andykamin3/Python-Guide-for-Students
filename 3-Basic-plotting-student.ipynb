## Examples
We will demonstrate the usage of `matplotlib` below. You will mostly work with the `plot` and `scatter` command. You can use the `title`, `xlabel`, `ylabel` and other commands to nicely format your plots. 

```python
# Basic example of plotting a sine wave using NumPy and Matplotlib
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 6 * np.pi, 500)
y = np.sin(t)

plt.plot(t, y)
plt.title("sin(t)")
plt.xlabel("t")
plt.ylabel("sin(t)")
plt.grid(True)
plt.show()
```

```python
# From 2-Matrices-and-vectors.ipynb
def update_dynamics(current_position_x, velocity_x, dt):
    # Simple dynamics update: new_position = old_position + velocity * time_step
    new_position_x = current_position_x + velocity_x * dt
    return new_position_x

x0 = np.array([0.0, 0.0,0.0])  # Initial position 2D
v = np.array([1.0, 0.5,0.0])   # Velocity in x and y
dt = 0.1                    # Time step in seconds

t_end = 1.0
N = int(t_end / dt)
positions = np.zeros((N, 3))  # To store positions at each time
times = np.linspace(0, t_end, N)  # Time array for plotting
positions[0] = x0

for i in range(N):
    positions[i] = update_dynamics(positions[i-1], v, dt)
```
