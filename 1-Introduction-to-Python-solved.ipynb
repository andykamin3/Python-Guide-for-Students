### Mission
The Crazyflie's motors have a maximum current limit to protect the battery and motor electronics. In this task, you will use a conditional to ensure that the commanded motor current does not exceed the `max_motor_current`. If the calculated current (`desired_current`) exceeds this limit, you should cap it to the maximum allowed value. This is a safety feature used in real drones.

```python
max_motor_current = 2.0  # [Amperes] - Maximum allowed motor current
desired_current = 2.5    # [Amperes] - Current command we want to send

print(f"Max Motor Current: {max_motor_current} [A]")
print(f"Desired Current: {desired_current} [A]")

# Limit the current to the maximum if it exceeds the limit
actual_current = None

if desired_current > max_motor_current:
    actual_current = max_motor_current
else:
    actual_current = desired_current

print(f"Actual Motor Current: {actual_current} [A]")
```

# Functions 
Functions are reusable blocks of code that perform specific tasks. They take inputs (parameters) and return outputs, creating a mapping between input and output values.
We use functions to make our code easier to understand and more modular. It can also help us avoid repeating code. 
In Python the keyword `def` is used to define a function, followed by the name 

```python
# Test the safety check functions with the existing max_motor_current variable
def is_current_safe(desired_current, max_current):
    if desired_current > max_current:
        return False
    else:
        return True
def show_current_safety(desired_current, max_motor_current):
    if is_current_safe(desired_current, max_motor_current):
        print("Current is within safe limits.")
    else:
        print("Warning: Desired current exceeds maximum limit!")


desired_current_1 = 1.5  # [Amperes] - A safe current command
desired_current_2 = 2.5  # [Amperes] - An unsafe current command

show_current_safety(desired_current_1, max_motor_current) # Expected output: "Current is within safe limits."
show_current_safety(desired_current_2, max_motor_current) # Expected output: "Warning: Desired current exceeds maximum limit!"
```

### Mission
Now it is time for you to define your own function. We will slowly start working on our crazyflie simulation. We will start by creating a function `update_dynamics` that updates the position of our robot based on the parameters:

- `current_position_x`: The current position in the `x` axis.
- `velocity_x`: The currrent velocity in the `x` axis.
- `dt`: The simulation timestep. 

The function should return:

- `new_position_x`: A number representing the new position in the `x` axis.




```python
def update_dynamics(current_position_x, velocity_x, dt):
    # Simple dynamics update: new_position = old_position + velocity * time_step
    new_position_x = current_position_x + velocity_x * dt
    return new_position_x


x0 = 0.0  # Initial position in meters
velocity_x = 2.0  # Velocity in meters per second
dt = 1.0  # Time step in seconds

x1 = update_dynamics(x0, velocity_x, dt)  # Position after one time step

print(f"Initial Position: {x0} m")
print(f"Velocity: {velocity_x} m/s")
print(f"Time Step: {dt} s")
print(f"Updated Position: {x1} m") # Expected output: "Updated Position: 2.0 m"
```

## Loops

Loops are used to repeat actions efficiently.

- A **`for` loop** is best when the number of iterations is known.
- A **`while` loop** is best when repeating until a condition becomes false.

### Mission
Let's use the `update_dynamics` function we created last time (remember that if you couldn't solve it you can check the `1-Introduction-to-Python-SOLVED.ipynb` notebook for a hint). This time we will reduce the `dt` variable to `0.1` and use a `for` loop to simulate `1.0` seconds into the future and update the dynamics by adding velocity.

```python
# Simulate 1.0 s into the future using smaller time step
dt = 0.1
simulation_time = 1.0
N = int(simulation_time / dt)

current_position_x = x0

for step in range(N):
    current_position_x = update_dynamics(current_position_x, velocity_x, dt)

print(f"Initial position: {x0:.2f} m")
print(f"Velocity: {velocity_x:.2f} m/s")
print(f"dt: {dt:.2f} s, steps: {N}")
print(f"Position after {simulation_time:.2f} s: {current_position_x:.2f} m")
```

### Mission
What if we wanted to use a `while` loop in this case? Now instead of using a `for` loop we want the robot to move until it has reached `x1 = 1.0` and print the time it took for the robot to reach that position.

```python
# Simulate until the position is 1m
current_position_x = x0
t = 0.0

while current_position_x < 1.0:
    current_position_x = update_dynamics(current_position_x, velocity_x, dt)
    t += dt

print(f"Initial position: {x0:.2f} m")
print(f"Velocity: {velocity_x:.2f} m/s")
print(f"dt: {dt:.2f} s")
print(f"Simulated time: {t:.2f} s")
print(f"Position after {t:.2f} s: {current_position_x:.2f} m")
```

Some packages like `math` are already included with Python. Below you can find examples of importing a module and using its methods. See that the `sin` and `cos` methods take radians as input (like in MATLAB).

```python
import math

drone_orientation_deg = 90
drone_orientation_rad = math.radians(drone_orientation_deg)
# Simple example: sine and cosine of the drone orientation
sin_orientation = math.sin(drone_orientation_rad)
cos_orientation = math.cos(drone_orientation_rad)

print(f"Orientation: {drone_orientation_deg}° ({drone_orientation_rad:.2f} rad)")
print(f"sin(theta) = {sin_orientation:.2f}")
print(f"cos(theta) = {cos_orientation:.2f}")
```
