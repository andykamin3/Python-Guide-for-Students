## Examples
We will demonstrate the usage of `matplotlib` below. You will mostly work with the `plot` and `scatter` command. You can use the `title`, `xlabel`, `ylabel` and other commands to nicely format your plots. 

```python
# Basic example of plotting a sine wave using NumPy and Matplotlib
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 6 * np.pi, 500)
y = np.sin(t)

plt.plot(t, y)
plt.title("sin(t)")
plt.xlabel("t")
plt.ylabel("sin(t)")
plt.grid(True)
plt.show()
```

### Mission
We will use the code from lab 2 to simulate a drone flying for one second with a given velocity. 
You will have to make the following plots:
1. $x(t)$
2. $y(t)$
3. $x-y$ projection. Add a red dot for the start position and a green one for the end position.

Use the `plot` method. Make sure to label your axis and add a title to your plot. 

```python
# From 2-Matrices-and-vectors.ipynb
def update_dynamics(current_position_x, velocity_x, dt):
    # Simple dynamics update: new_position = old_position + velocity * time_step
    new_position_x = current_position_x + velocity_x * dt
    return new_position_x

x0 = np.array([0.0, 0.0,0.0])  # Initial position 2D
v = np.array([1.0, 0.5,0.0])   # Velocity in x and y
dt = 0.1                    # Time step in seconds

t_end = 1.0
N = int(t_end / dt)
positions = np.zeros((N, 3))  # To store positions at each time
times = np.linspace(0, t_end, N)  # Time array for plotting
positions[0] = x0

for i in range(N):
    positions[i] = update_dynamics(positions[i-1], v, dt)
```

```python
# Plotting the trajectory in the x-axis as a function of time
plt.figure()
plt.plot(times, positions[:, 0], label='x position')
plt.title("Position over Time")
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.grid(True)
plt.legend()
plt.show()  

# Plotting the trajectory in the y-axis as a function of time
plt.figure()
plt.plot(times, positions[:, 1], label='y position', color='orange')
plt.title("Position over Time")
plt.xlabel("Time (s)")
plt.ylabel("Position (m)")
plt.grid(True)
plt.legend()    

# Plotting the trajectory in the xy-plane 
plt.figure()
plt.plot(positions[:, 0], positions[:, 1], label='Trajectory')    
plt.plot(positions[0, 0], positions[0, 1], 'ro', label='Start position')
plt.plot(positions[-1, 0], positions[-1, 1], 'go', label='End position')  
plt.title("Trajectory in XY-plane")
plt.xlabel("X Position (m)")
plt.ylabel("Y Position (m)")
plt.grid(True)
plt.legend()
plt.axis('equal')  # Set equal aspect ratio for x and y axes
plt.show()
```
